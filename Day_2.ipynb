```python
# Checking if the year is leap or not\

year=int(input("Enter the year:"))
if(year%100==0 and year%400==0):
  print("It's a leap year")
elif(year%100!=0 and year%4==0):
  print("It's a leap year")
else:
  print("It's not a leap year")
```
