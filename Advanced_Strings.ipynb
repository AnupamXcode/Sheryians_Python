```python
# Concept of Falsey Values
falsy_values = [
    False,
    None,
    0,
    0.0,
    0j,
    "",
    [],
    (),
    {},
    set(),
    frozenset()
]

for value in falsy_values:
    print(f"{repr(value):>12} -> {bool(value)}")
```

```python
# Strings

# Indexing
s = "Anshu"
print(s[0])    # A
print(s[-1])   # u
```

```python
#Slicing
s = "Programming"

print(s[0:4])     # Prog
print(s[:6])      # Progra
print(s[3:])      # gramming
print(s[-4:])     # ming
print(s[::2])     # Pormig
print(s[::-1])    # gnimmargorP (reverse)
```

```python
# Concatenation
a = "Hello"
b = "World"

print(a + " " + b)
print(a * 3)
```

```python
# Membership Operators
s = "Python"

print("Py" in s)       # True
print("Java" not in s)
print("f" in s)
```

```python
s =input("Enter the String:")

print("The length of String is :",len(s))

for ch in s:
    print(ch)
```

```python
# Case Conversion Methods
s = "pYtHon"

print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
print(s.swapcase())
```

```python
# WhiteSpace Methods
s = "  hello world  "

print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
a="Anupam Shrivastava"
print(a[5:17:1])
## [Start : End : Steps]
```

```python
# Searching and Counting
s = "banana"

print(s.find("a"))      # index or -1
print(s.rfind("a"))
print(s.index("n"))     # error if not found
print(s.count("a"))
```

```python
# Replace and translate
s = "I like Java"

print(s.replace("Java", "Python"))

table = str.maketrans("aeiou", "12345")
print("education".translate(table))
```

```python
# Split and Join
s = "one,two,three"

parts = s.split(",")
print(parts)

print("-".join(parts))
```

```python
# String Validation
s = "Python123"

print(s.isalpha())
print(s.isdigit())
print(s.isalnum())
print(s.islower())
print(s.isupper())
print(s.isspace())
```

```python
# Starts and Ends with
s = "file.txt"

print(s.startswith("file"))
print(s.endswith(".txt"))
```

```python
# String Formatting
name = "Anshu"
age = 20

print(f"My name is {name} and I am {age}")

print("Hello {}, age {}".format(name, age))
print("Hello %s" % name)
```

```python
# Escape Char
print("Line1\nLine2")
print("Tab\tSpace")
print("Quote \"Hello\"")
```

```python
# Encoding and Decoding

s = "hello"

encoded = s.encode("utf-8")
decoded = encoded.decode("utf-8")

print(encoded)
print(decoded)
```
