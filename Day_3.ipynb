```python
# Factorial of a number
n = int(input("Enter a no.: "))

if (n < 0):
    print("Factorial does not exist for negative numbers.")
elif (n == 0 or n == 1):
    print(1)
else:
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i  # Multiply the current number to the total
    print(fact)

```

```python
# Print the sum of all even & odd numbers in a range separately
n = int(input("Enter a number: "))
sum_even = 0
sum_odd = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        sum_even += i  # Sirf even numbers yahan add honge
    else:
        sum_odd += i   # Baaki (odd) numbers yahan add honge

print(f"The Sum of Even numbers is: {sum_even}")
print(f"The Sum of Odd numbers is: {sum_odd}")
```

```python
#Accept a number and check if it a perfect number or not. A number whose sum of factors is equal to the number itself
#Ex - 6 = 1, 2, 3 =
n = int(input("Enter a number: "))
sum_factors = 0

# Hum 1 se n-1 tak loop chalayenge (number khud factor nahi hona chahiye)
for i in range(1, n):
    if n % i == 0:
        sum_factors += i

if sum_factors == n:
    print(f"{n} is a Perfect Number.")
else:
    print(f"{n} is not a Perfect Number.")
```

```python
# Check wether the number is prime or not
import math
def check_prime(n):
    if n <= 1: return False
    # Sirf square root tak check karein
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
print("Prime" if check_prime(num) else "Not Prime")
```

```python
str1 = "P@#yn26at^&i5ve"

chars = 0
digits = 0
symbols = 0

for char in str1:
    if char.isalpha():
        chars += 1
    elif char.isdigit():
        digits += 1
    else:
        symbols += 1

print(f"Chars = {chars}")
print(f"Digits = {digits}")
print(f"Symbol = {symbols}")
```
